```python
# import necessary libraries
import pandas as pd
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv("/content/sample_data/sample_text.csv")
df.shape
```

```python
df
```

```python
from sentence_transformers import SentenceTransformer

encoder=SentenceTransformer("all-mpnet-base-v2")
vectors=encoder.encode(df.text)
vectors.shape
```

```python
dim=vectors.shape[1]
dim
```

```python
import faiss

index=faiss.IndexFlatL2(dim)
```

```python
index.add(vectors)
```

```python
search_query="I wnat to visit Japan"
svec=encoder.encode(search_query)
svec.shape
```

```python
import numpy as np


svec=np.array(svec).reshape(1,-1)
svec.shape
```

```python
distance,I=index.search(svec,k=2)
```

```python
df.loc[I[0]]
```
